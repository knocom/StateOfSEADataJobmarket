# I. INDEED: JOB SCRAPING

```python
#regex library for use later on and pandas for converting extracted list data to DataFrame
import re
import pandas as pd
```

```python
#function to remove \nnew from Job Title
def strip_newline(list_):
    list2_ = []
    for element in list_:
        if element[-4:]=='\nnew':
            element = element.strip('\nnew')
            list2_.append(element)
        else:
            list2_.append(element)
    list_ = list2_
    return list_
```

```python
#Combined Scrape for Indeed's Page 1 and 2 (~100 results; 50 results per page) roughly 1hr and 15mins runtime
#Selenium requires installed chromedriver.exe installed in a directory (if using Google Chrome to scrape)
import selenium
import time
import random
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from fake_useragent import UserAgent

options = Options()
ua = UserAgent()
userAgent = ua.random #get random fake user agent details
print(userAgent)
options.add_argument(f'user-agent={userAgent}')

driver = webdriver.Chrome(options=options, executable_path=r"C:\Program Files (x86)\chromedriver.exe")
driver.get('https://ph.indeed.com/jobs?q=business+intelligence&limit=50&radius=100&start=100') ##Enter Target URL here. This is #101-150 for BI jobs in the PH.
time.sleep(90) #sometimes need extra time to do a reCaptcha challenge
joblist = [] #create empty list for job titles
companylist = [] #create empty list for company names
locationlist = [] #create empty list for locations
urllist = [] #create empty list for the URLs of the postings
jobdesclist = [] #create empty list for the job descriptions

job_titles = driver.find_elements_by_class_name('title')
for job in job_titles:
    print(job.text)
    joblist.append(job.text)
    job.click() #need to retrieve URL for each job title by clicking on it
    time.sleep(0.5)
    urllist.append(driver.current_url)  #retrieves current URL and appends it to the urllist
time.sleep(3)
companies = driver.find_elements_by_class_name('company')
for company in companies:
    print(company.text)
    companylist.append(company.text)
time.sleep(3)
locations = driver.find_elements_by_class_name('location')
for location in locations:
    print(location.text)
    locationlist.append(location.text)
time.sleep(3)

print(joblist)
print(companylist)
print(locationlist)
print(urllist)


for i in range(len(urllist)):
    driver.get(urllist[i])
    time.sleep(20)
    try:
        driver.find_element_by_xpath('//*[@id="popover-x"]').click() #in case a popup appears, it should click X
        time.sleep(10)
        job_description = driver.find_element_by_xpath('//*[@id="vjs-desc"]')
        jobdesclist.append(job_description.text)
        print(i)
        print(jobdesclist[i])
        time.sleep(random.randint(7,10))
    except:
        try:
            job_description = driver.find_element_by_xpath('//*[@id="vjs-desc"]') #this is for when no popup appears
            jobdesclist.append(job_description.text)
            print(i)
            print(jobdesclist[i])
            time.sleep(random.randint(7,10))
        except:
            try:
                job_description = driver.find_element_by_xpath('//*[@id="jobDescriptionText"]') #sometimes it opens a new window instead of opening the sidebar, which changes the element we have to find
                jobdesclist.append(job_description.text)
                print(i)
                print(jobdesclist[i])
                driver.back() #go back a page on your browser
                time.sleep(random.randint(15,23))
            except:
                continue

print('\n')
print(jobdesclist)

driver.get('https://ph.indeed.com/jobs?q=business+intelligence&limit=50&radius=100&start=150') #Enter Target URL here. This is #151-200 of BI jobs in PH.
time.sleep(90)  #sometimes need extra time to do a reCaptcha challenge
joblist2 = []
companylist2 = []
locationlist2 = []
urllist2 = []
jobdesclist2 = []

job_titles2 = driver.find_elements_by_class_name('title')
for job in job_titles2:
    print(job.text)
    joblist2.append(job.text)
    job.click() #need to retrieve URL for each job title by clicking on it
    time.sleep(0.5)
    urllist2.append(driver.current_url)  #retrieves current URL and appends it to the urllist
time.sleep(3)
companies2 = driver.find_elements_by_class_name('company')
for company in companies2:
    print(company.text)
    companylist2.append(company.text)
time.sleep(3)
locations2 = driver.find_elements_by_class_name('location')
for location in locations2:
    print(location.text)
    locationlist2.append(location.text)
time.sleep(3)

print(joblist2)
print(companylist2)
print(locationlist2)
print(urllist2)


for i in range(len(urllist2)):
    driver.get(urllist2[i])
    time.sleep(20)
    try:
        driver.find_element_by_xpath('//*[@id="popover-x"]').click() #in case a popup appears, it should click X
        time.sleep(5)
        job_description2 = driver.find_element_by_xpath('//*[@id="vjs-desc"]')
        jobdesclist2.append(job_description2.text)
        print(i)
        print(jobdesclist2[i])
        time.sleep(random.randint(5,7))
    except:
        try:
            job_description2 = driver.find_element_by_xpath('//*[@id="vjs-desc"]') #this is for when no popup appears
            jobdesclist2.append(job_description2.text)
            print(i)
            print(jobdesclist2[i])
            time.sleep(random.randint(5,7))
        except:
            try:
                job_description2 = driver.find_element_by_xpath('//*[@id="jobDescriptionText"]') #sometimes it opens a new window instead of opening the sidebar, which changes the element we have to find
                jobdesclist2.append(job_description2.text)
                print(i)
                print(jobdesclist2[i])
                driver.back()
                time.sleep(random.randint(15,23))
            except:
                continue

print('\n')
print(jobdesclist2)
```

```python
import selenium
import time
import random
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from fake_useragent import UserAgent

options = Options()
ua = UserAgent()
userAgent = ua.random
#print(userAgent)
options.add_argument(f'user-agent={userAgent}')
driver = webdriver.Chrome(options=options, executable_path=r"C:\Program Files (x86)\chromedriver.exe")
#driver.get('https://www.glassdoor.com/Job/jobs.htm?suggestCount=0&suggestChosen=false&clickSource=searchBtn&typedKeyword=&locT=N&locId=204&jobType=&context=Jobs&sc.keyword=Data+Scientist&dropdown=0')
#https://www.glassdoor.com/Job/philippines-data-scientist-jobs-SRCH_IL.0,11_IN204_KO12,26.htm //page1
driver.get('https://www.glassdoor.com/Job/jobs.htm?suggestCount=0&suggestChosen=false&clickSource=searchBtn&typedKeyword=data+an&locT=N&locId=113&jobType=&context=Jobs&sc.keyword=Data+Analyst&dropdown=0')
time.sleep(90)
joblist = []
urllist = []
jobs = driver.find_elements_by_css_selector("a.jobInfoItem.jobTitle.css-13w0lq6.eigr9kq1.jobLink")
for job in jobs:
    print(job.text)
    joblist.append(job.text)
    urllist.append(job.get_attribute('href'))
    
companylist = []
companies = driver.find_elements_by_css_selector("a.css-10l5u4p.e1n63ojh0.jobLink")
for company in companies:
    print(company.text)
    companylist.append(company.text)
    
locationlist = []
locations = driver.find_elements_by_css_selector("span.loc.css-nq3w9f.pr-xxsm")
for location in locations:
    print(location.text)
    locationlist.append(location.text)
    
agelist = []
ages = driver.find_elements_by_css_selector("div.d-flex.align-items-end.pl-std.css-mi55ob")
for age in ages:
    print(age.text)
    agelist.append(age.text)
    
descriptionlist = []
jobinsightlist = []
industrylist = []
sizelist = []
employmenttypelist = []
jobfunctionlist = []

for url in urllist:
    driver.get(url)
    time.sleep(10)
    description = driver.find_element_by_css_selector("div.desc.css-58vpdc.ecgq1xb4")
    if description != '':
        print(description.text)
        descriptionlist.append(description.text)
        #css-u9c4ai e18tf5om1
        #css-15asexb e18tf5om0
        #css-15asexb e18tf5om0
    insight = driver.find_element_by_css_selector("div.css-15asexb.e18tf5om0")
    if insight != '':
        print(insight.text)

        employmenttype = insight.text.splitlines()[2]
        print(employmenttype)
        
    
        jobfunction = insight.text.splitlines()[4]
        print(jobfunction)
        
    
        industry = insight.text.splitlines()[6]
        print(industry)
        
    
        size = insight.text.splitlines()[8]
        print(size)
        
    else:
        employmenttpye = 'N/A'
        jobfunction = 'N/A'
        industry = 'N/A'
        size = 'N/A'
        
    employmenttypelist.append(employmenttype)
    jobfunctionlist.append(jobfunction)
    industrylist.append(industry)
    sizelist.append(size)
    
#function to remove \nnew from Job Title
def strip_newline(list_):
    list2_ = []
    for element in list_:
        if element[-4:]=='\nnew':
            element = element.strip('\nnew')
            list2_.append(element)
        else:
            list2_.append(element)
    list_ = list2_
    return list_

stripdesclist = strip_newline(descriptionlist)
stripdesclist

zippedList = list(zip(joblist,companylist,locationlist,agelist,stripdesclist,industrylist,sizelist,employmenttypelist,jobfunctionlist,urllist))

import pandas as pd
dfGlassdoor = pd.DataFrame(zippedList, columns = ['JobTitle','Company','Location','Age','JobDescription','Industry','Size','EmploymentType','JobFunction','Url'])

import re
dfGlassdoor['JobDescription'] = dfGlassdoor['JobDescription'].replace(r'\n',' ',regex=True)

#save to csv 
dfGlassdoor.to_csv('C:/Users/Janina/Downloads/GlassdoorID_DA1.csv') 
```
