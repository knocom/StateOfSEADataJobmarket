# Applying Doccano made Customer NER on raw csv file

```python
import numpy as np
import pandas as pd
import spacy
from spacy.pipeline import EntityRecognizer
```

```python
df = pd.read_csv('df_combined_all_raw_v2.csv', encoding='latin1', index_col='Index')
#have to make encoding latin1 because it gives an error if reading in utf-8
```

```python
df
```

```python
df.columns
```

```python
df.shape
```

```python
nlp = spacy.load('en_core_web_md', disable=['ner'])
ner = EntityRecognizer(nlp.vocab)
ner.from_disk('./spacytrain/to/ner')
nlp.add_pipe(ner,"custom_ner")
print(nlp.meta['pipeline'])
```

```python
doc=nlp(u'We are seeking a Data Scientist skilled in machine learning and algorithms. Should be knowledgeable on Python, R and SQL. Must have 10-15 years of experience and a graduate of a quantitative degree such as a Bachelor or Master in Statistics, IT, or Mathematics.')
for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
from spacy import displacy
displacy.serve(doc, style="ent")
```

```python
def get_spec_skill(text):
    doc = nlp(text)
    return [x.text for x in doc.ents if x.label_=='SPEC_SKILL']

def get_gen_skill(text):
    doc = nlp(text)
    return [x.text for x in doc.ents if x.label_=='GEN_SKILL']

def get_yrs_of_exp(text):
    doc = nlp(text)
    return [x.text for x in doc.ents if x.label_=='YRS_OF_EXP']

def get_educ_level(text):
    doc = nlp(text)
    return [x.text for x in doc.ents if x.label_=='EDUCATION_LVL']

def get_degree(text):
    doc = nlp(text)
    return [x.text for x in doc.ents if x.label_=='DEGREE']    

def get_salary(text):
    doc = nlp(text)
    return [x.text for x in doc.ents if x.label_=='SALARY'] 
```

```python
#this will take awhile maybe around 30mins
df['Data Skills - Specific'] = df['Job Description'].apply(get_spec_skill)
df['Data Skills - General'] = df['Job Description'].apply(get_gen_skill)
df['Experience (years)'] = df['Job Description'].apply(get_yrs_of_exp)
df['Education Level'] = df['Job Description'].apply(get_educ_level)
df['Degree'] = df['Job Description'].apply(get_degree)
df['Salary'] = df['Job Description'].apply(get_salary)
```

```python
df = pd.read_csv('df_EskwelabsCapstonev2.csv', index_col='Index')
```

```python
import ast
```

```python
#Cleaning data by converting string values to list type and removing duplicates

#Data Skills - Specific
result = []
for value in df['Data Skills - Specific']:
    try:
        result.append(ast.literal_eval(value)) #this will convert a string representation of a list into a real list
    except:
        result.append('') #for empty or NaN values
df['Data Skills - Specific'] = result

result2 = []
for value in df['Data Skills - Specific']:
    mylist = value
    mylist = list(dict.fromkeys(mylist)) #removes duplicates in a list by converting it into a dict (which can't have duplicate keys) then converting it back to a list
    result2.append(mylist)
df['Data Skills - Specific'] = result2

#Data Skills - General

result = []
for value in df['Data Skills - General']:
    try:
        result.append(ast.literal_eval(value)) 
    except:
        result.append('')
df['Data Skills - General'] = result

result2 = []
for value in df['Data Skills - General']:
    mylist = value
    mylist = list(dict.fromkeys(mylist))
    result2.append(mylist)
df['Data Skills - General'] = result2

#Data Skills - Experience (years)

result = []
for value in df['Experience (years)']:
    try:
        result.append(ast.literal_eval(value)) 
    except:
        result.append('')
df['Experience (years)'] = result

result2 = []
for value in df['Experience (years)']:
    mylist = value
    mylist = list(dict.fromkeys(mylist))
    result2.append(mylist)
df['Experience (years)'] = result2

#Data Skills - Education Level

result = []
for value in df['Education Level']:
    try:
        result.append(ast.literal_eval(value)) 
    except:
        result.append('')
df['Education Level'] = result

result2 = []
for value in df['Education Level']:
    mylist = value
    mylist = list(dict.fromkeys(mylist))
    result2.append(mylist)
df['Education Level'] = result2

#Data Skills - Degree

result = []
for value in df['Degree']:
    try:
        result.append(ast.literal_eval(value)) 
    except:
        result.append('')
df['Degree'] = result

result2 = []
for value in df['Degree']:
    mylist = value
    mylist = list(dict.fromkeys(mylist))
    result2.append(mylist)
df['Degree'] = result2

#Data Skills - Salary

result = []
for value in df['Salary']:
    try:
        result.append(ast.literal_eval(value)) 
    except:
        result.append('')
df['Salary'] = result

result2 = []
for value in df['Salary']:
    mylist = value
    mylist = list(dict.fromkeys(mylist))
    result2.append(mylist)
df['Salary'] = result2
```
